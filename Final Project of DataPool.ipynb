## Mobile Sales

```python
import pandas as p
```

```python
df=p.read_csv('mobile_sales.csv')
```

```python
print(df)
```

## Describe the components of the dataset

```python
print(df.describe())
```

## Which Brand has the highest priced product?

```python
print(df[['TransactionID','Brand','Price']].max())
```

## Which Brand has the minimum priced product?

```python
print(df[['TransactionID','Brand','Price']].min())
```

## Display the payment methods cash for different brands

```python
print(df.query('PaymentMethod == "Cash"')[['Brand', 'PaymentMethod']])
```

## Display the number of payments made through different methods

```python
print(df[['Brand', 'PaymentMethod']].groupby('PaymentMethod').size())
```

## How many customers are there whose age is greater than 60 i.e senior citizens who are buying mobile phones of which brands?

```python
print(df.query('CustomerAge > 60')[['Brand', 'CustomerAge']])
```
